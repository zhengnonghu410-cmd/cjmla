# 🍄 Super Mario - Build Android APK

点击菜单栏 **Runtime → Run all**，等待约30分钟即可下载 APK。

```
# 1. 安装系统依赖
!sudo apt update -qq
!sudo apt install -y -qq python3-pip python3-dev openjdk-17-jdk-headless zip unzip autoconf libtool pkg-config zlib1g-dev libncurses5-dev libncursesw5-dev cmake libffi-dev libssl-dev libltdl-dev
```

```
# 2. 安装 Buildozer
!pip install -q buildozer "cython<3.0"
```

```
# 3. 从 GitHub 下载项目代码
!rm -rf super-mario
!git clone https://github.com/zhengnonghu410-cmd/cjmla.git super-mario
%cd super-mario
```

```
# 4. 创建 buildozer.spec 配置文件
%%writefile buildozer.spec
[app]
title = Super Mario
package.name = supermario
package.domain = org.supermario
source.dir = .
source.include_exts = py,png,jpg,gif,json,ogg,wav
source.exclude_dirs = .idea, __pycache__, .git, .github
version = 1.0
requirements = python3==3.10.11,pygame
orientation = landscape
fullscreen = 1
android.permissions = INTERNET
android.api = 33
android.minapi = 21
android.ndk = 25b
android.accept_sdk_license = True
p4a.branch = develop
p4a.bootstrap = sdl2

[buildozer]
log_level = 2
warn_on_root = 1
```

```
# 5. 开始构建 APK（约30分钟）
!buildozer android debug
```

```
# 6. 下载 APK
from google.colab import files
import glob
apks = glob.glob('bin/*.apk')
if apks:
    files.download(apks[0])
    print(f"✅ APK 已下载: {apks[0]}")
else:
    print("❌ 没有找到 APK 文件，请检查上面的构建日志")
```
